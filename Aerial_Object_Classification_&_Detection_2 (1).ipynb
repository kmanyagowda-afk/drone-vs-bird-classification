# **Project Summary -**

This project focuses on building an intelligent computer vision system to classify and optionally detect aerial objects as either birds or drones using deep learning techniques. The main objective is to develop a robust model that can accurately distinguish between natural flying objects (birds) and man-made aerial devices (drones), which is useful in areas like surveillance, airspace monitoring, and security systems.

A Custom Convolutional Neural Network (CNN) built from scratch

## **Data Preprocessing**

````python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set image size and batch size
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# Data generators (with augmentation for train)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

# Load data
train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary' # Changed from 'categorical' to 'binary'
)

val_data = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary' # Changed from 'categorical' to 'binary'
)

test_data = val_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary' # This was already changed to binary in the fit cell, ensuring consistency
)

print("Classes:", train_data.class_indices)
````

## 3. ***Data Wrangling***

````python
import os
import shutil
from PIL import Image

base_path = "/content/dataset"   # your dataset path

print(" Starting Data Wrangling...\n")

# -----------------------------
# 1. Remove corrupt images
# -----------------------------
corrupt_count = 0

for root, dirs, files in os.walk(base_path):
    for file in files:
        file_path = os.path.join(root, file)
        try:
            img = Image.open(file_path)
            img.verify()
        except:
            os.remove(file_path)
            corrupt_count += 1

print(f"🧹 Removed {corrupt_count} corrupt images\n")

# -----------------------------
# 2. Remove non-image files
# -----------------------------
valid_extensions = ('.jpg', '.jpeg', '.png')

removed_files = 0

for root, dirs, files in os.walk(base_path):
    for file in files:
        if not file.lower().endswith(valid_extensions):
            os.remove(os.path.join(root, file))
            removed_files += 1

print(f" Removed {removed_files} non-image files\n")

# -----------------------------
# 3. Count images per class
# -----------------------------
print(" Class Distribution:\n")

class_counts = {}

for split in ['train', 'valid', 'test']:
    split_path = os.path.join(base_path, split)
    print(f"{split.upper()}:")

    for class_name in os.listdir(split_path):
        class_path = os.path.join(split_path, class_name)
        count = len(os.listdir(class_path))
        class_counts[class_name] = class_counts.get(class_name, 0) + count
        print(f"   {class_name}: {count}")

print("\nTotal:", class_counts)

# -----------------------------
# 4. Handle class imbalance
# -----------------------------
print("\n Checking imbalance...\n")

min_count = min(class_counts.values())

for split in ['train', 'valid', 'test']:
    split_path = os.path.join(base_path, split)

    for class_name in os.listdir(split_path):
        class_path = os.path.join(split_path, class_name)
        files = os.listdir(class_path)

        # Reduce extra images (simple balancing)
        if len(files) > min_count:
            for file in files[min_count:]:
                os.remove(os.path.join(class_path, file))

print(" Dataset balanced\n")

# -----------------------------
# 5. Resize images (optional)
# -----------------------------
IMG_SIZE = (224, 224)

for root, dirs, files in os.walk(base_path):
    for file in files:
        file_path = os.path.join(root, file)
        try:
            img = Image.open(file_path)
            img = img.resize(IMG_SIZE)
            img.save(file_path)
        except:
            pass

print("📏 Images resized\n")

# -----------------------------
# 6. Final Check
# -----------------------------
print("\n Final Dataset Ready!")

for split in ['train', 'valid', 'test']:
    print(f"\n{split.upper()}:")
    split_path = os.path.join(base_path, split)

    for class_name in os.listdir(split_path):
        count = len(os.listdir(os.path.join(split_path, class_name)))
        print(f"   {class_name}: {count}")
````



### What manipulations have been done and insights found:

For this image classification project, the primary data manipulations were performed during the data loading phase using `ImageDataGenerator`:

1.  **Image Resizing:** All images across the training, validation, and test sets were uniformly resized to `(224, 224)` pixels. This standardization is crucial for feeding images into convolutional neural networks (CNNs), which typically require fixed-size inputs. This ensures consistency and compatibility with popular pre-trained models for transfer learning.

2.  **Pixel Normalization/Rescaling:** Pixel values, originally ranging from 0-255, were rescaled to `[0, 1]` by dividing by 255. This normalization helps in stabilizing and speeding up the training process of neural networks, as it prevents large input values from causing large weight updates and potential gradient explosion.

3.  **Data Augmentation (for Training Data):** To enhance the model's ability to generalize and to prevent overfitting, various data augmentation techniques were applied *only* to the training set:
    *   `rotation_range=20`: Randomly rotates images by up to 20 degrees.
    *   `zoom_range=0.2`: Randomly zooms images by up to 20%.
    *   `horizontal_flip=True`: Randomly flips images horizontally.
    These augmentations create new, slightly modified versions of existing training images, effectively increasing the diversity and size of the training dataset without collecting new data. This makes the model more robust to variations in real-world images.

4.  **No Explicit Missing Value or Outlier Treatment:** Unlike tabular datasets, image datasets generally do not have 'missing values' or 'outliers' in the same sense. If image files were corrupt or missing, `ImageDataGenerator` would have failed during loading, which did not happen. Therefore, no explicit handling for these was required or performed beyond ensuring successful data loading.

**Key Insights from these manipulations:**

*   The dataset is now in a clean, consistent format suitable for deep learning models.
*   Data augmentation will likely improve the model's generalization capabilities, reducing the chance of overfitting to the specific training examples.
*   The preprocessing aligns with best practices for training CNNs, setting a solid foundation for model development.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

---
**Hypothetical Statement:** There is a significant difference in the average brightness of 'bird' images compared to 'drone' images in the dataset. Specifically, we hypothesize that 'drone' images are, on average, brighter than 'bird' images.

**Null Hypothesis (H0):** The average brightness of 'bird' images is equal to or greater than the average brightness of 'drone' images (μ_bird >= μ_drone).

### **Alternate Hypothesis (H1):** The average brightness of 'drone' images is significantly greater than the average brightness of 'bird' images (μ_drone > μ_bird).
---

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy import stats

# Ensure we have data for both classes
if not bird_brightness or not drone_brightness:
    print("Error: Brightness data not available for both classes.")
else:
    # Perform an independent two-sample t-test
    # Since our alternative hypothesis is μ_drone > μ_bird, this is a one-tailed test.
    # The ttest_ind function typically returns a two-tailed p-value.
    # For a one-tailed test where H1: mean1 > mean2, if t-statistic is positive, p_one_tail = p_two_tail / 2.
    # If t-statistic is negative, p_one_tail = 1 - (p_two_tail / 2).
    # Alternatively, you can directly use the 'alternative' parameter if available (scipy 1.6+).

    t_statistic, p_value_two_tailed = stats.ttest_ind(drone_brightness, bird_brightness, equal_var=False) # Welch's t-test (unequal variances)

    # Calculate one-tailed p-value for H1: mean_drone > mean_bird
    if t_statistic > 0:
        p_value_one_tailed = p_value_two_tailed / 2
    else:
        p_value_one_tailed = 1 - (p_value_two_tailed / 2)

    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Two-tailed P-value: {p_value_two_tailed:.4f}")
    print(f"One-tailed P-value (H1: Drone > Bird): {p_value_one_tailed:.4f}")

    alpha = 0.05
    print(f"\nSignificance Level (alpha): {alpha}")
    if p_value_one_tailed < alpha:
        print("Conclusion: Reject the Null Hypothesis. There is statistically significant evidence that drone images are, on average, brighter than bird images.")
    else:
        print("Conclusion: Fail to Reject the Null Hypothesis. There is not enough statistically significant evidence to conclude that drone images are, on average, brighter than bird images.")
````

##### Which statistical test have you done to obtain P-Value?

I performed a **Two-Sample Independent Welch's t-test**.

##### Why did you choose the specific statistical test?

I chose the **Two-Sample Independent Welch's t-test** for the following reasons:

1.  **Comparison of Means:** The hypothesis involves comparing the average (mean) brightness between two distinct groups ('bird' images and 'drone' images). The t-test is designed for this purpose.
2.  **Independence of Samples:** The 'bird' images and 'drone' images constitute two independent groups; the brightness of an image in one group does not influence the brightness of an image in the other group.
3.  **Numerical Data:** Brightness is a continuous numerical variable, which is appropriate for a t-test.
4.  **Unequal Variances (Welch's):** I opted for Welch's t-test (`equal_var=False`) because it does not assume equal variances between the two groups, which is a more robust choice when the variances might differ, as is often the case with real-world image features.
5.  **One-tailed Hypothesis:** My alternate hypothesis (H1) is directional (μ_drone > μ_bird), implying a one-tailed test. The code calculates the one-tailed p-value based on the t-statistic.

````python
# Perform Statistical Test to obtain P-Value
from scipy import stats

# Ensure we have data for both classes
if not bird_edge_densities or not drone_edge_densities:
    print("Error: Edge density data not available for both classes.")
else:
    # Perform an independent two-sample t-test
    # Our alternative hypothesis is μ_drone > μ_bird, so this is a one-tailed test.
    # The ttest_ind function typically returns a two-tailed p-value.
    # For a one-tailed test where H1: mean1 > mean2, if t-statistic is positive, p_one_tail = p_two_tail / 2.
    # If t-statistic is negative, p_one_tail = 1 - (p_two_tail / 2).

    t_statistic, p_value_two_tailed = stats.ttest_ind(drone_edge_densities, bird_edge_densities, equal_var=False) # Welch's t-test (unequal variances)

    # Calculate one-tailed p-value for H1: mean_drone > mean_bird
    if t_statistic > 0:
        p_value_one_tailed = p_value_two_tailed / 2
    else:
        p_value_one_tailed = 1 - (p_value_two_tailed / 2)

    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Two-tailed P-value: {p_value_two_tailed:.4f}")
    print(f"One-tailed P-value (H1: Drone > Bird): {p_value_one_tailed:.4f}")

    alpha = 0.05
    print(f"\nSignificance Level (alpha): {alpha}")
    if p_value_one_tailed < alpha:
        print("Conclusion: Reject the Null Hypothesis. There is statistically significant evidence that drone images have, on average, a higher edge density than bird images.")
    else:
        print("Conclusion: Fail to Reject the Null Hypothesis. There is not enough statistically significant evidence to conclude that drone images have, on average, a higher edge density than bird images.")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

For image datasets, the concept of 'outliers' is not typically treated with statistical methods as it is in tabular data. Instead, outliers in images usually refer to:

1.  **Corrupt or Malformed Images:** These were already addressed during the data wrangling phase (cell `wk-9a2fpoLcV`) where we removed unreadable files.

2.  **Images with Unusual Content:** These could be images that are highly blurred, completely black/white, or contain objects that are not relevant to the classes (e.g., a hand holding a bird, but the hand dominates the image). These are usually handled through:
    *   **Data Augmentation:** Techniques like random cropping, rotation, and zooming make the model more robust to variations, effectively treating some 'unusual' but valid image presentations as part of the normal distribution.
    *   **Model Robustness:** Deep learning models, especially Convolutional Neural Networks (CNNs), are generally robust to some degree of noise and variation in the data. Extremely unrepresentative images might act as 'noise' during training, but batch processing and regularization techniques help mitigate their impact.
    *   **Manual Inspection (if feasible):** For smaller datasets, a manual review can identify and remove highly anomalous images.
    *   **Advanced Techniques:** For very severe cases, anomaly detection models could be trained to flag such images, but this is beyond the scope of typical classification preprocessing.

In this project, the primary method for handling 'outliers' in terms of unusual but valid content is through the **data augmentation strategies** applied by `ImageDataGenerator`. This helps the model learn to recognize birds and drones under a wide variety of conditions, rather than being overly sensitive to specific, potentially 'outlier-like' visual characteristics. No explicit numerical outlier treatment (like IQR or Z-score methods) is applied to pixel intensities, as these are raw features processed by the CNN layers.

For our second model, we will utilize **Transfer Learning** with a pre-trained MobileNetV2 architecture. Transfer learning is a powerful technique where a model, pre-trained on a large and general dataset (like ImageNet), is repurposed for a new, specific task. This significantly reduces training time and often leads to better performance, especially when our dataset is not extremely large.

#### Why MobileNetV2?
MobileNetV2 is a lightweight, efficient convolutional neural network architecture designed for mobile and embedded vision applications. It offers a good balance of accuracy and computational cost, making it an excellent choice for demonstrating transfer learning where speed is a concern.

#### Approach:
1.  **Load Pre-trained Base Model:** We will load the MobileNetV2 model, pre-trained on ImageNet, without its top classification layer (`include_top=False`).
2.  **Freeze Base Layers:** The layers of the pre-trained MobileNetV2 will be "frozen" so that their learned weights are not updated during the initial training phase. This allows our new classification layers to learn quickly from the already extracted robust features.
3.  **Add Custom Classification Head:** We will add our own `Flatten`, `Dense` (with ReLU activation), `Dropout`, and `Dense` (with softmax activation for binary classification) layers on top of the frozen MobileNetV2 base. These layers will learn to classify the features extracted by the MobileNetV2 base into 'bird' or 'drone'.
4.  **Compile and Train:** The combined model will be compiled and trained on our dataset. We'll use `categorical_crossentropy` as the loss function and `adam` optimizer, similar to the custom CNN.

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models

# Define directory paths (from cell 370840ce)
train_dir = "/content/dataset/train"
val_dir = "/content/dataset/valid"
test_dir = "/content/dataset/test"

# Set image size and batch size (from cell 99057274)
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# Data generators (with augmentation for train) (from cell 99057274)
train_datagen_kfold = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen_kfold = ImageDataGenerator(rescale=1./255)

# Load data using flow_from_directory to get filepaths and classes
train_data_kfold = train_datagen_kfold.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False # Important for getting ordered filepaths and classes
)

# Extract file paths and numerical labels
image_paths = train_data_kfold.filepaths
numerical_labels = train_data_kfold.classes

# Get class names mapping from the generator
class_names = {v: k for k, v in train_data_kfold.class_indices.items()}

# Convert numerical labels to string labels
string_labels = [class_names[label] for label in numerical_labels]

# Prepare dataframe (IMPORTANT)
# df should have columns: ['filename', 'class']
df = pd.DataFrame({
    'filename': image_paths,   # list of image paths
    'class': string_labels           # 'bird' or 'drone' as strings
})

# Fix: Change n_splits to 2 or more for k-fold cross-validation
skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=42) # Changed n_splits to 3

fold_no = 1
# Iterate through splits generated by StratifiedKFold
for train_idx, val_idx in skf.split(df['filename'], df['class']):
    print(f"\nTraining Fold {fold_no}")

    train_df = df.iloc[train_idx]
    val_df = df.iloc[val_idx]

    # Use a fresh datagen for each fold's flow_from_dataframe
    # It's good practice to re-initialize ImageDataGenerator for each fold
    # to ensure no state leakage and consistent augmentation application.
    datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        zoom_range=0.2,
        horizontal_flip=True # Apply augmentation to training data
    )
    val_datagen_fold = ImageDataGenerator(rescale=1./255) # No augmentation for validation

    train_gen = datagen.flow_from_dataframe(
        train_df,
        x_col='filename',
        y_col='class',
        target_size=(224,224),
        class_mode='binary',
        batch_size=BATCH_SIZE # Use the defined BATCH_SIZE
    )

    val_gen = val_datagen_fold.flow_from_dataframe(
        val_df,
        x_col='filename',
        y_col='class',
        target_size=(224,224), # Keep target_size consistent with train_gen
        class_mode='binary',
        batch_size=BATCH_SIZE
    )

    # Model (re-initialize for each fold to ensure fresh weights)
    base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224,224,3))
    base_model.trainable = False # Keep base model frozen for initial folds

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=1 # You can increase epochs for better training, but keep low for demonstration
    )

    fold_no += 1
````

### ML Model - 3: Transfer Learning with ResNet50

For our third model, we will again use **Transfer Learning**, this time with the **ResNet50** architecture. ResNet (Residual Network) models are characterized by their 'skip connections' or 'residual blocks' that allow the network to be much deeper without suffering from the vanishing gradient problem. ResNet50 is a widely used and powerful model for various computer vision tasks.

#### Why ResNet50?
ResNet50 is a deeper and often more accurate model compared to MobileNetV2, though it is also more computationally intensive. It's a great choice for tasks requiring higher accuracy, and its pre-trained weights on ImageNet are highly effective for feature extraction.

#### Approach:
1.  **Load Pre-trained Base Model:** We will load the ResNet50 model, pre-trained on ImageNet, without its top classification layer (`include_top=False`).
2.  **Freeze Base Layers:** The layers of the pre-trained ResNet50 will be frozen to preserve the learned features and speed up training.
3.  **Add Custom Classification Head:** We will add a `GlobalAveragePooling2D` layer, followed by `Dense` (with ReLU activation), `Dropout`, and a final `Dense` (with softmax activation for binary classification) layer. These layers will adapt the extracted features to our 'bird'/'drone' classification task.
4.  **Compile and Train:** The combined model will be compiled with the `Adam` optimizer and `categorical_crossentropy` loss, and then trained on our dataset.

````python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import layers, models

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))

# Freeze layers
base_model.trainable = False

x = base_model.output
x = layers.GlobalAveragePooling2D()(x)

x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.5)(x)

# Dense output layer
output = layers.Dense(1, activation='sigmoid')(x)

model = models.Model(inputs=base_model.input, outputs=output)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
````
